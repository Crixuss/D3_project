```python
import pandas as pd
import numpy as np
```

```python
beer_list = pd.read_csv("beers.csv", encoding="latin-1")
brewery_list = pd.read_csv("breweries.csv", encoding = "latin-1")
```

```python
beer_list.head()
```

```python
brewery_list.head()
```

```python
brewery_list.rename(columns={'Unnamed: 0':'brewery_id'}, 
                 inplace=True)
brewery_list.head()
```

```python
print(beer_list.dtypes)
print(brewery_list.dtypes)
```

```python
beer_merged_list = pd.merge(beer_list,brewery_list, how = "left", on="brewery_id", sort=True)
beer_merged_list
```

```python
beer_merged_list.rename(columns={
                        'Unnamed: 0':'unknown',
                        'abv':'beer_abv',
                        'ibu':'beer_ibu',
                        'id':'beer_id',
                        'name_x':'beer_name',
                        'style':'beer_style',
                        'brewery_id':'brewery_id',
                        'ounces':'beer_ounces',
                        'name_y':'brewery_name',
                        'city':'brewery_city',
                        'state':'brewery_state',
}, 
                 inplace=True)
beer_merged_list.head()
```

```python
cleaned_beer_list = beer_merged_list.drop(columns=['unknown'])
cleaned_beer_list
```

```python
len(cleaned_beer_list) - cleaned_beer_list['beer_ibu'].count()
```

```python
len(cleaned_beer_list) - cleaned_beer_list['beer_abv'].count()
```

```python
beer_list_nonulls = cleaned_beer_list.dropna()
beer_list_nonulls
```

```python
beer_list_nonulls.dtypes
```

```python
import config
from sqlalchemy import create_engine
beer_data = f'postgres:{config.password}@localhost:5432/beer_db'
engine = create_engine(f'postgresql://{beer_data}')
```

```python
engine.table_names()
```

```python
beer_list_nonulls.to_sql(name='beer_data', con=engine, if_exists='append', index=False)
pd.read_sql_query('select * from beer_data', con=engine).head()
```
